```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori , association_rules
```

```python
df=pd.read_csv('Oder3.csv')
```

```python
df
```

```python
transactions = df.groupby(['TransactionNo'])['Items'].apply(list).values.tolist()
```

```python
transactions
```

```python
te = TransactionEncoder()
one_hot_encoded = te.fit(transactions).transform(transactions)
one_hot_encoded
```

```python
ndf = pd.DataFrame(one_hot_encoded,columns=te.columns_)
```

```python
ndf
```

```python
frequent_itemsets = apriori(ndf,min_support=0.01,use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=0.5)
```
