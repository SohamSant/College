```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cities_r2.csv")
```

```python
X1= df[['effective_literacy_rate_total','population_total']]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X1_scaled = sc.fit_transform(X1)
```

```python
from sklearn.cluster import KMeans

model =KMeans(n_clusters=3,random_state=0)

df['Cluster']=model.fit_predict(X1_scaled)
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='effective_literacy_rate_total',y='population_total',hue='Cluster',data=df,s=200)
plt.title("K-Means")
plt.xlabel('Literacy Rate')
plt.ylabel('Total Population')
plt.show()
```
