```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cities_r2.csv')
```

```python
df
```

```python
X1  = df[['total_graduates', 'population_total']]
```

```python
X1
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X1_scaled = sc.fit_transform(X1)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3,random_state=0)
df['Cluster']=model.fit_predict(X1_scaled)


```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_graduates', y='population_total', hue='Cluster', data=df,  s=100)
plt.title('K-Means Clustering of Customers based on Total Graduates and Population Total')
plt.xlabel('Total Graduates')
plt.ylabel('Population Total')

plt.show()
```
